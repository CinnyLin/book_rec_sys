```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('clustering_output4.csv', index_col=0)
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
df['predict_cluster_kmeans'].value_counts()
```

```python
plt.bar(df['predict_cluster_kmeans'].value_counts().keys(), list(df['predict_cluster_kmeans'].value_counts()), color='#5e99c5')
plt.xticks(np.arange(0,3,1))
plt.show()
```

```python
for col in df.columns:
    plt.figure(figsize = (5,5))
    ax = sns.barplot(x = 'predict_cluster_kmeans', y = col, data = df,
                     palette = ['#5e99c5'], 
                     linewidth = 1.5, edgecolor = 'black')
    plt.xlabel('')
    plt.ylabel('')
    plt.title(f'{col}')
    #plt.savefig(f'cluster_plots/{col}.png')
```

```python
plt.figure(figsize = (5,5))
ax = sns.barplot(x = 'predict_cluster_kmeans', y = 'rating', data = df,
                 palette = ['lightsteelblue', '#5e99c5', '#2a6a9e'], 
                 linewidth = 1.5, edgecolor = 'black')
plt.xlabel('')
plt.ylabel('')
plt.title('User Rating', fontsize=10)
plt.show()
```
